```python
data = pd.read_csv('Shanghai_HMT_2010.csv')
```

```python
data = data.dropna()
```

```python
print("data.shape = ", data.shape)
data.head(10)
```

```python
def func(x):
    if x == 'NE':
        return 1
    elif x  == 'NW':
        return 2
    elif x == 'SW':
        return -1
    elif x == 'SW':
        return -2        
    else:
        return 3

data['cbwd'] = data.cbwd.apply(func)

#data['cbwd'].apply(lambda x: 1 if x == 'NE'  2 elif x == 'NW'  -1 elif == 'SW'  -2 elif == 'SE'  else 3)
```

```python
print("data.shape = ", data.shape)
data.head(10)
```

```python
a = data['cbwd']
unique_numbers = list(set(a))
print(unique_numbers)
```

```python
data = (data - data.mean()) / data.std()
```

```python
print("data.shape = ", data.shape)
data.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[['No',	'year',	'month',	'day',	'hour',	'season',	'PM_Jingan',	'PM_US Post',	'PM_Xuhui',	'DEWP',	'HUMI', 'TEMP',	'cbwd',	'Iws',	'precipitation',	'Iprec']], data['PRES'], test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, explained_variance_score
```

```python
model = LinearRegression(normalize=True)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```
