```python
import math as m
import numpy as np
import matplotlib.pyplot as plt


x_d = np.linspace(0, 2*m.pi, 13)
y_d = np.cos(0.67 * x_d)
plt.plot(x_d,y_d)
plt.show()
```

```python
y_noise = 0.2 * np.random.normal(size = x_d.size)
print(y_noise)
```

```python
y_dd = y_d + y_noise
```

```python
def ccos(x, p):
  return(np.cos(p*x))
```

```python
import scipy as sp
import scipy.optimize
from scipy.optimize import curve_fit
```

```python
p_sup, er = curve_fit(ccos, x_d, y_dd)
print(p_sup, er)
```

```python
y = lambda x: np.cos(p_sup* x)
x = np.linspace(0, 2*m.pi,100)
plt.plot(x, y(x))
```

```python
plt.plot(x, y(x), label = 'функция')
plt.plot(x_d, y_d, label = 'набор данных')

plt.legend()
plt.show()
```
